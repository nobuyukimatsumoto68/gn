```python
import numpy as np
import matplotlib.pyplot as plt

import scipy
from scipy import special as sp

import glob
```

```python
# beta=2.0
Nc=3
```

```python
basedir="obs_exc/"
```

```python
filelist = sorted(glob.glob(basedir+"trPhisq*")[:-1])
```

```python
data_=[]
for f in filelist:
    dat = np.loadtxt( f )
    data_.append(dat)
data=np.array(data_)
```

```python
data.shape
```

```python
for a in np.arange(0, data.shape[0], 5):
    plt.plot(data[a].T[1], alpha=0.5, label=str(data[a][0][0]))

plt.xlabel("MCT")
plt.ylabel("$\\vert \\Phi-1 \\vert^2$")
plt.legend(loc="upper right")
plt.title("$SU(3)$, single plaquette, trivial $G$")

plt.show()
```

```python
for a in np.arange(0, data.shape[0], 4):
    dat = data[a]
    plt.hist(dat.T[1]/Nc**2,
             density=True,
             alpha=0.2, bins=20, label=str(data[a][0][0]))
plt.legend()

plt.xlabel("$\\vert \\Phi-1 \\vert^2$")

plt.show()
```
